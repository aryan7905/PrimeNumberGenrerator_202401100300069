```python
import random

def is_prime(n):
    """Check if a number is prime."""
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False

    i = 5
    while i * i <= n:
        if n % i == 0 or n % (i + 2) == 0:
            return False
        i += 6

    return True

def generate_random_prime(min_val=2, max_val=100):
    """Generate a random prime number within the given range."""
    while True:
        num = random.randint(min_val, max_val)
        if is_prime(num):
            return num
```

```python
# Cell 2: Generate a random prime number
# Run this cell to generate a new random prime

min_range = 2
max_range = 1000
random_prime = generate_random_prime(min_range, max_range)

print(f"Generated prime number: {random_prime}")
```

```python
# Run this cell whenever you want to check a number

number_to_check = int(input("Enter a number to check if it's prime: "))
result = is_prime(number_to_check)

print(f"{number_to_check} is {'prime' if result else 'not prime'}")
```

```python
# Visualize prime numbers on a graph
import matplotlib.pyplot as plt
import numpy as np

def get_primes_up_to(limit):
    """Get all prime numbers up to the given limit."""
    primes = []
    for num in range(2, limit + 1):
        if is_prime(num):
            primes.append(num)
    return primes

# Set the limit for visualization
limit = 100

# Get all primes up to the limit
primes = get_primes_up_to(limit)

# Create data for visualization
all_numbers = list(range(2, limit + 1))
is_prime_list = [1 if is_prime(n) else 0 for n in all_numbers]

# Create the plot
plt.figure(figsize=(12, 6))

# Plot 1: Distribution of prime numbers
plt.subplot(1, 2, 1)
plt.scatter(all_numbers, is_prime_list, color='blue', alpha=0.7)
plt.xlabel('Number')
plt.ylabel('Is Prime (1=Yes, 0=No)')
plt.title(f'Distribution of Prime Numbers up to {limit}')
plt.grid(True, alpha=0.3)

# Plot 2: Count of primes vs. non-primes
plt.subplot(1, 2, 2)
plt.plot(all_numbers, np.cumsum(is_prime_list), color='red', label='Cumulative Count of Primes')
plt.xlabel('Number')
plt.ylabel('Count of Primes')
plt.title(f'Growth of Prime Numbers up to {limit}')
plt.grid(True, alpha=0.3)
plt.legend()

plt.tight_layout()
plt.show()

# Print some statistics
print(f"Total numbers checked: {len(all_numbers)}")
print(f"Total prime numbers found: {len(primes)}")
print(f"Percentage of primes: {(len(primes) / len(all_numbers)) * 100:.2f}%")
print(f"Prime numbers found: {primes}")
```
